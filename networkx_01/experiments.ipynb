### Here, we will create a synthetic street network using NetworkX:

#### - Nodes represent intersections or endpoints of streets.

#### - Edges represent the streets themselves.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a simple 5x5 grid graph
G = nx.grid_2d_graph(5, 5)
pos = {node: node for node in G.nodes()}  # Use node coordinates as positions

# Visualize the grid
nx.draw(G, pos, with_labels=True, node_size=300, node_color="lightblue")
plt.title("Grid Graph (Synthetic Street Network)")
plt.show()
```

### we can also make a real-world example, to extract a street network froma specific location. It will be done using OSMnx.

### This graph represents real-world streets and intersections, making it highly relevant for Street-Based Tessellation.

```python
import osmnx as ox

# Download street network of a real place
G = ox.graph_from_place("Jaipur, India", network_type="drive")

# Visualize the network
ox.plot_graph(G)
```
